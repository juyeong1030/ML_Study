```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
iris_data = load_iris()
```

```python
knn_model = KNeighborsClassifier()
```

```python
X = pd.DataFrame(data = iris_data['data'], columns = iris_data['feature_names'])
```

```python
y = iris_data['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 6, test_size = 0.3)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
knn_model.fit(X_train, y_train)
```

```python
knn_model.score(X_train, y_train)
```

```python
knn_model.score(X_test, y_test)
```

```python
pd.plotting.scatter_matrix(X, figsize = (10, 10), alpha = 0.8, c = y)
plt.show()
```
