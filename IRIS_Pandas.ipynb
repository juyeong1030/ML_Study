```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns = iris.feature_names)
df_iris['target'] = iris.target
```

```python
df_iris = pd.DataFrame(iris['data'], columns = iris['feature_names']).assign(target = iris['target'])

target_converter = {idx : target for idx, target in enumerate(iris['target_names'])}
df_iris['target_name'] = df_iris.target.map(target_converter)
df_iris
```

```python
df_iris.describe()
```

```python
df_iris['color'] = np.where(df_iris.target_name == 'setosa', 'red', np.where(df_iris.target_name == 'versicolor', 'green', 'blue'))

df_iris.plot.scatter(x = 'petal length (cm)', y = 'petal width (cm)', s = 50, alpha = 0.5, c = df_iris['color'])
plt.show()
```

```python
plt.figure(figsize = (5,5))
sns.scatterplot(x = 'petal length (cm)', y = 'petal width (cm)', hue = 'target_name', palette = 'viridis', data = df_iris)
plt.show()
```
